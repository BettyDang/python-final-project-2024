```python
import pyodbc

# Connection for Windows Authentication
connection_string = (
    "Driver={ODBC Driver 17 for SQL Server};"
    "Server=BETTY;"
    "Database=Pokedex;"
    "Trusted_Connection=yes;"
)

try:
    conn = pyodbc.connect(connection_string)
    print("Connection successful!  Printing...")

    cursor = conn.cursor()

    query = "SELECT ID, name, type FROM Pokemon"
    cursor.execute(query)

    rows = cursor.fetchall()
    for row in rows:
        print(f"{row.ID}, {row.name}, Type: {row.type}")

except Exception as e:
    print("Error connecting to the database:", e)

finally:
    try:
        cursor.close()
    except:
        pass
    try:
        conn.close()
    except:
        pass

print("End of First Generation")
```
